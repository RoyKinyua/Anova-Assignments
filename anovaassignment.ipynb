```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
```

#### 1. Marketing Campaign Effectiveness

```python
campaign_A = [12, 15, 14, 10, 13, 15, 11, 14, 13, 16]
campaign_B = [18, 17, 16, 15, 20, 19, 18, 16, 17, 19]
campaign_C = [10, 9, 11, 10, 12, 9, 11, 8, 10, 9]

f_stat, p_val = f_oneway(campaign_A, campaign_B, campaign_C)
print(f"F-statistic: {f_stat:.2f}, p-value: {p_val:.5f}")
```

##### null hypothesis: There is no difference between the means of the three groups
##### alternative hypothesis: There is a difference between the means of the three groups

##### Conclusion: We will reject the null hypothesis because the p-value is less than the alpha. This means the three campaigns are statistically significant so we would reject the null hypothesis

#### 3.Teaching Method VS Student Performance

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
data = {
    'score': [78, 85, 80, 90, 88, 82, 84, 86, 92],
    'method': ['Lecture', 'Lecture', 'Lecture',
               'Discussion', 'Discussion', 'Discussion',
               'Project', 'Project', 'Project']
}
```

```python
df =pd.DataFrame(data)


model = ols('score ~ C(method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

#### null hypothesis: There is no difference between the means of the students score therefore concluding the different teaching techniques did not work
#### alternative hypothesis: There is a difference between the means of the students exam scores thus concluding the different teaching techniques worked
#### conclusion: The data is not statistically significant thus we do not reject the null hypothesis

#### 5. Diet Plan VS Weight Loss

```python
data = {
    'diet': ['Keto']*6 + ['Paleo']*6 + ['Vegan']*6 + ['Mediterranean']*6,
    'weight_loss': [6, 5, 7, 8, 6, 7, 5, 6, 4, 5, 6, 5, 3, 4, 2, 4, 3, 2, 5, 6, 7, 6, 8, 7]
} 

```

```python
df = pd.DataFrame(data)

model = ols('weight_loss ~ C(diet)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

#### null hypothesis: There is no difference between the mean weight loss thus concluding the different diet plans did not work.
#### alternative hypothesis: There is a difference between the mean weight loss therefore concluding the different diet plans worked.
#### conclusion: The p-value is less than alpha so we reject the null hypothesis
